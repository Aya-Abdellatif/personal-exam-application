```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
dataset = pd.read_csv("all_data.csv")
```

Visualizations

```python
#Bar char
topic_counts = dataset["topic"].value_counts()
plt.figure(figsize=(12, 8))
plt.bar(x = topic_counts.index, height = topic_counts, color = "midnightblue")
plt.xticks(rotation = 45,fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("SubTopic Distribution by Topic", fontsize = 16, fontweight = "bold")
plt.xlabel("Number of subTopics", fontsize = 13)
plt.show()
```
