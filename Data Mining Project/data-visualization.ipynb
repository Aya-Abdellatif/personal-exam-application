```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
dataset = pd.read_csv("all_data.csv")
```

Visualizations

```python
topic_counts = dataset["topic"].value_counts()
plt.figure(figsize=(12, 8))
plt.bar(x = topic_counts.index, height = topic_counts, color = "midnightblue")
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Question Distribution by Topic", fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Questions", fontsize = 13)
plt.show()
```

```python
subTopic_counts_per_topic = dataset.groupby('topic')['subTopic'].nunique()
plt.figure(figsize=(12, 8))
plt.bar(x = subTopic_counts_per_topic.index, height = subTopic_counts_per_topic, color = "midnightblue")
plt.title("Number of SubTopics per Topic", fontsize=16, fontweight="bold")
plt.ylabel("Number of SubTopics", fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```
